## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
# YOUR CODE HERE

W, R, B = 10, 12, 8
T = W + R + B
```

```python
p1 = W/T
print("P(White) =", round(p1,2))
```

```python
#White then black without replacement
p2 = (W/T)*(B/(T-1))
print("P(white then black, no replacement) =", round(p2,2))
```

```python
#White then black with replacement
p3 = (W/T)*(B/T)
print("P(white then black, with replacement) =", round(p3,2))
```

```python
#Favorable ways = W * R * B
favorable = W * R * B
#Total ways = number of ways to pick 3 from 30:
total = (T * (T - 1) * (T - 2)) / 6  # 30*29*28 / 6

p4 = favorable / total
print("P(3 balls all different) =", round(p4,2))
```

## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

Step 1 ‚Äî Define events

Let:
C = cloudy
R = rainy

Given:
ùëÉ(ùê∂)=0.40
P(R‚à£C)=0.50
P(R‚à£not C)=0.20

We want:
P(R)

Step 2 ‚Äî Use total probability formula

Rain can happen in two ways:
Case A: cloudy AND rainy
P(C)‚ãÖP(R‚à£C)=0.40‚ãÖ0.50=0.20

Case B: not cloudy AND rainy
1-P(C)‚ãÖP(R‚à£notC)=0.60‚ãÖ0.20=0.12

Step 3 ‚Äî Add them
P(R)=0.20+0.12=0.32

```python
# YOUR CODE HERE
P_cloudy = 0.40
P_rain_given_cloudy = 0.50
P_rain_given_not_cloudy = 0.20

P_rain = (
    P_cloudy * P_rain_given_cloudy +
    (1 - P_cloudy) * P_rain_given_not_cloudy
)

print(f"Chance of rain = {P_rain:.2f}")
print(f"Chance of rain = {P_rain*100:.0f}%")
```

## Challenge 4

 * If someone flies to China, the probability of getting Covid is 1%  
 * If someone flies to Italy, the probability of getting Covid is 20%  

I roll 1 dice with numbers 1, 2, 3, 4, 5, 6:  
 * If 1: I fly to Italy
 * If not 1: I fly to China

During my trip, I got Covid. What is the probability that I flew to China?

```python
# YOUR CODE HERE
P_china = 5/6
P_italy = 1/6

P_covid_given_china = 0.01
P_covid_given_italy = 0.20

# total probability of covid
P_covid = (
    P_china * P_covid_given_china +
    P_italy * P_covid_given_italy
)

# Bayes
P_china_given_covid = (P_covid_given_china * P_china) / P_covid

print(f"P(China | Covid) = {P_china_given_covid:.2f}")
print(f"= {P_china_given_covid*100:.0f}%")
```

## BONUS

You roll 2 dice. What is the probability that the sum of the numbers is 7? And 8? And 9? Then simulate 10000 times and draw a countplot.

your explanation here

```python
# YOUR CODE HERE
import random
from collections import Counter
import matplotlib.pyplot as plt
#Total trials
N = 10_000

# simulate sums
sums = []
for _ in range(N):
    d1 = random.randint(1, 6)
    d2 = random.randint(1, 6)
    sums.append(d1 + d2)

# count each sum
counts = Counter(sums)

print("Simulated probabilities:")
for s in [7, 8, 9]:
    print(f"Sum {s}: count={counts[s]}, prob={counts[s]/N:.4f}")

# countplot-like bar chart
x = list(range(2, 13))
y = [counts[i] for i in x]

plt.figure(figsize=(8, 4.5))
plt.bar(x, y)
plt.xticks(x)
plt.xlabel("Sum of two dice")
plt.ylabel("Count (out of 10,000)")
plt.title("Counts of sums from rolling 2 dice (10,000 trials)")
plt.tight_layout()
plt.show()
```
